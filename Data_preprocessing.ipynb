```python
df["date_closed"] = df["date_closed"].fillna("Not Closed")
df.isnull().sum()
```

```python
df["date_raised"] = pd.to_datetime(df["date_raised"])
df["date_closed"] = pd.to_datetime(
    df["date_closed"], errors="coerce"
)
```

```python
text_cols = ["query_heading", "query_description", "status"]

for col in text_cols:
    df[col] = df[col].str.lower()
```

```python
for col in text_cols:
    df[col] = df[col].str.strip()
```
